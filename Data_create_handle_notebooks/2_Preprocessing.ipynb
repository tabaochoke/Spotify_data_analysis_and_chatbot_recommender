- 'danceability' - Mức độ phù hợp của bài hát với việc nhảy múa trên thang tăng dần từ 0.0 tới 1.0, tức phù hợp trong những bữa tiệc.

- 'energy' - Độ năng lượng của bài hát từ 0.0 tới 1.0. Bài hát có năng lượng cao được xét dựa trên cường độ của âm, tempo và nhiều thứ khác. Ví dụ, nhạc rock thường được xem là có năng lượng mạnh, trong khi nhạc đồng quê được xem là nhẹ nhàng.

- 'loudness' - Cường độ âm thanh (theo thang dB).

- 'mode' - Chế độ là âm trưởng hoặc thứ.

- 'speechiness' - Mức độ hiện diện của giọng nói trên thang từ 0 tới 1. Ví dụ, podcast hoặc sách nói sẽ có speechiness cao, nhạc rap cũng mang xu hướng speechiness tương đối cao. Trong khi đó, những nhạc thông thường hoặc nhạc không lời sẽ có speechiness thấp.

- 'acousticness' - Tính âm thanh của nhạc cụ. Có nghĩa là âm thanh được ghi trong bài đến từ âm thanh của các nhạc cụ không phải là nhạc được sản xuất bởi các âm thanh điện tử.

- 'instrumentalness' - Tính nhạc cụ của bài hát. Tức là nếu bài hát có instrumentalness cao, nghĩa là ít vocal, tức là nhạc gốc không lời.

- 'liveness' - Mức độ hiện diện của khán giả trong bài nhạc. Nếu có tiếng khán giả, tức giá trị cao thì chứng tỏ bài nhạc được hát trực tiếp chứ không phải ghi âm.

- 'valence' - Giá trị tích cực với độ tích cực giảm dần từ 1.0 tới 0.0

- 'tempo' - Nhịp, tính bằng số nhịp (beats) trên phút (BPM). Tempo càng cao thì nhạc càng nhanh.

- 'type' - Loại của bài (Trong data tất cả là audio_features).

- 'id' - Mã số bài hát trong Spotify

- 'uri' - Định danh tài nguyên thống nhất

- 'track_href' - Endpoint API của bài hát

- 'analysis_url' - URL phân tích bài hát

- 'duration_ms' - Thời lượng bài hát (theo miligiây)

- 'time_signature' - Dấu thời gian. Tức số nhịp (beats) trên mỗi khung nhạc (bar)

- 'popularity' - Độ phổ biến của bài hát trên thang 100

- 'song_name' - Tên bài hát

Không có cột có kiểu dữ liệu sai. Released_date nên về dạng dictionary, sau đó trích ra cột released_year với kiểu dữ liệu int.


```python
def get_year(date_string):
    return int(date_string.split(":")[1].split(",")[0])
```

```python
df['released_year'] = df['release_date'].apply(lambda x: get_year(x) if not pd.isnull(x) else np.nan)
```

## 5) Phân bố data:

5.1) Data dạng số:

```python
num_col_info_df = df.select_dtypes(np.number)
def missing_ratio(input):
    return (input.isnull().sum() / len(input) * 100).round(1)
def lower_quartile(input):
    return input.dropna().quantile(0.25).round(1)
def median(input):
    return input.dropna().quantile(0.50).round(1)
def upper_quartile(input):
    return input.dropna().quantile(0.75).round(1)
num_col_info_df = num_col_info_df.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])
num_col_info_df
```

5.2) Data không phải dạng số:

```python
cat_col_info_df = df.select_dtypes(include='object')

def missing_ratio(input):
    return (input.isnull().sum() / len(input) * 100).round(1)
def num_values(input):
    return input.dropna().nunique()
def value_ratios(input):
    value_ratio = (input.value_counts() / len(input) * 100).round(1)
    return value_ratio.to_dict()

cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df
```

## 6) Xử lí cột 'lyric':

```python
from bs4 import BeautifulSoup
import re
```

```python
def lyric_cleanser(lyric):
  soup = BeautifulSoup(lyric, 'html.parser')
  clean_lyric = soup.get_text()
  clean_lyric = re.sub(r'\[.*?\]', '', clean_lyric)
  clean_lyric = re.sub(r'\n', ' ', clean_lyric)
  clean_lyric = re.sub(r'\s+', ' ', clean_lyric)
  return clean_lyric
```

```python
df['lyric'] = df['lyric'].apply(lambda x: lyric_cleanser(x))
```
